# Get kpi score

```python
month = input('Enter Month : ')
month_list = month.split(' ')

kpi_score = pd.read_csv('kpi_score.csv')

if len(month_list) == 1:
    kpi_score = kpi_score[['kpi_name', month_list[0]]]

elif len(month_list) > 1:
    kpi_score['avg_threshold'] = kpi_score[month_list].mean(axis=1)

kpi_score
```

```python
def kpi_score_func(x):
    col_list = list(x.columns)

    for col in col_list:
        if col in kpi_score.iloc[:,0].to_list():
            if len(month_list) == 1:
                x[col] = x[col].apply(lambda x: 100 if x >= kpi_score.loc[col, month_list[0]] else 0)
            else:
                return 'else'


kpi_score_func(df)
df
```
