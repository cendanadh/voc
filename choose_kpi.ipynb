# Import Library

```python
import pandas as pd
import numpy as np
```

# Import Data

```python
benchmark = pd.read_csv('data/voc_benchmark_monthly.csv')
user_experience = pd.read_csv('data/voc_user_experience_monthly.csv')
network_perfomance = pd.read_csv('data/voc_network_perfomance_monthly.csv')
```

# Get user input kpi list

```python
benchmark_list = list(benchmark.columns)
benchmark_list = [item for item in benchmark_list if item not in ['location_id', 'month', 'location', 'operator']]
benchmark_list = pd.DataFrame(benchmark_list,  columns=['kpi_name'])
benchmark_list['category'] = 'benchmark'

user_experience_list = list(user_experience.columns)
user_experience_list = [item for item in user_experience_list if item not in ['location_id', 'month', 'location', 'operator']]
user_experience_list = pd.DataFrame(user_experience_list,  columns=['kpi_name'])
user_experience_list['category'] = 'user_experience'

network_perfomance_list = list(network_perfomance.columns)
network_perfomance_list = [item for item in network_perfomance_list if item not in ['location_id', 'month', 'location', 'operator']]
network_perfomance_list = pd.DataFrame(network_perfomance_list,  columns=['kpi_name'])
network_perfomance_list['category'] = 'network_perfomance'

kpi_list = pd.concat([benchmark_list, user_experience_list, network_perfomance_list], axis=0, ignore_index=True)
kpi_list
```

```python
benchmark_select = input('Select KPI index from Benchmark Group : ')
benchmark_select = benchmark_select.split(' ')
benchmark_select = list(map(int, benchmark_select))
benchmark_select = [item for item in benchmark_select if item in range(0, 26)]
benchmark_select = list(set(benchmark_select))
benchmark_select.sort()

user_experience_select = input('Select KPI index from User Experience Group : ')
user_experience_select = user_experience_select.split(' ')
user_experience_select = list(map(int, user_experience_select))
user_experience_select = [item for item in user_experience_select if item in range(26, 35)]
user_experience_select = list(set(user_experience_select))
user_experience_select.sort()

network_perfomance_select = input('Select KPI index from Network Perfomance Group : ')
network_perfomance_select = network_perfomance_select.split(' ')
network_perfomance_select = list(map(int, network_perfomance_select))
network_perfomance_select = [item for item in network_perfomance_select if item in range(35, 46)]
network_perfomance_select = list(set(network_perfomance_select))
network_perfomance_select.sort()
```

```python
selected_kpi = pd.DataFrame()
selected_kpi = selected_kpi = pd.concat([selected_kpi, benchmark.iloc[:,[0,1,2]]], axis=1)

for x in benchmark_select:
    col_name = kpi_list.iloc[x, 0]
    selected_kpi = pd.concat([selected_kpi, benchmark.loc[:,col_name]], axis=1)

for x in user_experience_select:
    col_name = kpi_list.iloc[x, 0]
    selected_kpi = selected_kpi.merge(user_experience.loc[:,['location_id', 'month', col_name]], on=['location_id', 'month'], how='left')

for x in network_perfomance_select:
    col_name = kpi_list.iloc[x, 0]
    selected_kpi = selected_kpi.merge(network_perfomance.loc[:,['location_id', 'month', col_name]], on=['location_id', 'month'], how='left')
```

# Get user input month

```python
try:
    month = input('Enter Month : ')
    month_list = month.split(' ')
    month_list = list(map(int, month_list))

    for x in month_list:
        if x < 1 or x > 12:
            print('Error')
            del month_list
            break
        else:
            continue
except:
    del month_list
    print("Error")

month_list
```

```python
if len(month_list) > 1:
    selected_kpi = selected_kpi[selected_kpi['month'].isin(month_list)]
else:
    selected_kpi = selected_kpi[selected_kpi['month'] == month_list[0]]
    
#selected_kpi = selected_kpi.reset_index()
selected_kpi
```

# Aggregate per month

```python
selected_kpi_agg = selected_kpi.groupby(['location_id', 'location']).mean()
del selected_kpi_agg['month']
selected_kpi_agg = selected_kpi_agg.reset_index()
selected_kpi_agg
```

# Export to csv

```python
selected_kpi.to_csv('selected_kpi.csv')
selected_kpi_agg.to_csv('selected_kpi_agg.csv')
```
